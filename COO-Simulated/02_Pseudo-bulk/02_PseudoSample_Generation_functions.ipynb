Function to generate pseudo-samples from downsampled scRNA-Seq data. 
This is normalised to the median of total counts of cells, aggregated together, and CPM normalised.


```python
import os
import scanpy as sc
import pandas as pd
import numpy as np

def generate_pseudo_samples(ref_path: str, output_file: str):
    """
    Function to generate pseudo-samples from scRNA-Seq data in the given H5AD file(s) 
    and store all results in one output file.
    
    Args:
        ref_path (str): Path to the input H5AD file(s).
        output_file (str): Path to the output text file where all results will be saved.
    """
    # List all h5ad files in the current directory
    if os.path.isdir(ref_path):
        files = [f for f in os.listdir(ref_path) if f.endswith('.h5ad')]
    else:
        files = [ref_path]

    # Initialize an empty DataFrame to store all results
    all_results = pd.DataFrame()

    for file in files:
        # Read the h5ad file
        adata = sc.read_h5ad(os.path.join(ref_path, file) if os.path.isdir(ref_path) else file)
        
        # Use 'decontXcounts' layer for analysis
        adata.X = adata.layers["decontXcounts"]
        
        # Remove dissociation genes


        # Normalize the data (step 1)
        sc.pp.normalize_total(adata)  # This normalizes the data by the total count per cell
        
        # Sum all cells (columns) on the normalized data (step 2)
        summed_data = adata.X.sum(axis=0).A1  # .A1 converts sparse matrix to array
        
        # Ensure summed_data is not empty
        if summed_data.size == 0:
            print(f"Warning: No data to sum in {file}. Skipping...")
            continue
        
        # CPM normalization (step 3)
        total_sum = summed_data.sum()  # Total sum of counts across all genes
        if total_sum == 0:
            print(f"Warning: Total sum is zero for {file}. Skipping CPM normalization...")
            continue
        
        cpm_data = (summed_data / total_sum) * 1e6  # CPM normalization
        
        # Create a DataFrame for the summed and normalized data (transpose the data)
        summed_df = pd.DataFrame(cpm_data, columns=[file], index=adata.var_names)
        
        # Concatenate the result to the all_results DataFrame
        all_results = pd.concat([all_results, summed_df], axis=1)

        print(f"Processed {file} and added to the results.")

    # Fill NaN values with 0
    all_results.fillna(0, inplace=True)

    # Convert all values to floats as CIBERSORTx decon does not handle this
    all_results = all_results.astype(float)

    # Save the resulting DataFrame as a single output txt file
    all_results.to_csv(output_file, sep="\t")

    print(f"All pseudo-samples saved to {output_file}")

## Usage for V1 samples
#generate_pseudo_samples('COO-Simulated/02_Pseudo-bulk/Uniform_v1', 'COO-Simulated/02_Pseudo-bulk/Uniform_v1/Uniform_v1_All-Counts.txt')
#generate_pseudo_samples('COO-Simulated/02_Pseudo-bulk/Random_v1', 'COO-Simulated/02_Pseudo-bulk/Random_v1/Random_v1_All-Counts.txt')

# Usage for V2 samples
generate_pseudo_samples('COO-Simulated/02_Pseudo-bulk/Random_v2C', 'COO-Simulated/02_Pseudo-bulk/Random_v2C_New_New_New_New_New_New_New_New_New/Random_v2C_All-Counts.txt')
```
