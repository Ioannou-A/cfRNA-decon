Function to generate pseudo-samples from downsampled scRNA-Seq data. 
This is normalised to the median of total counts of cells, aggregated together, and CPM normalised.


```python
import os
import pandas as pd

def merge_proportion_files(directory, output_file):
    """
    Function to merge all files ending in *_subsampling_proportions.txt from a given directory.
    The function ensures that all unique columns from different files are included, 
    filling missing values with 0.
    
    Args:
        directory (str): Path to the directory containing the input files.
        output_file (str): Path to the output text file where the merged results will be saved.
    """
    # Get all files matching the pattern *_subsampling_proportions.txt
    files = [f for f in os.listdir(directory) if f.endswith("_subsampling_proportions.txt")]
    
    if not files:
        print("No matching files found.")
        return
    
    dataframes = []
    
    # Read all files into dataframes
    for file in files:
        filepath = os.path.join(directory, file)
        df = pd.read_csv(filepath, sep='\t')  # Assuming tab-separated files
        dataframes.append(df)
    
    # Merge all dataframes with outer join, filling missing values with 0
    merged_df = pd.concat(dataframes, join='outer', ignore_index=True).fillna(0)
    
    # Save the merged dataframe to a file
    merged_df.to_csv(output_file, sep='\t', index=False)
    print(f"Merged file saved as {output_file}")


merge_proportion_files('COO-Simulated/02_Pseudo-bulk/Uniform_v1', 'COO-Simulated/02_Pseudo-bulk/Uniform_v1/Uniform_v1_All-Proportions.txt')
merge_proportion_files('COO-Simulated/02_Pseudo-bulk/Random_v1', 'COO-Simulated/02_Pseudo-bulk/Random_v1/Random_v1_All-Proportions.txt')
```
