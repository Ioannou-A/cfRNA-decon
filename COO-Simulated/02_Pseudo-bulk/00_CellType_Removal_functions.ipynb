```python
# Function to remove the cell types that are absent from the custom matrices (TSP-HBA and TSP-BDa)

import os
import scanpy as sc

def filter_h5ad_by_cell_type(ref_path, out_dir, cell_types_of_interest, ct_key="cell_ontology_class"):
    """
    Filters an h5ad file by keeping only specific cell types.

    Parameters:
    ref_path (str): Path to the input h5ad file.
    out_dir (str): Directory to save the filtered h5ad file.
    cell_types_of_interest (list): List of cell types to keep.
    ct_key (str): Key in .obs that contains cell type annotations.

    Returns:
    str: Path to the filtered h5ad file.
    """
    
    # Read the input AnnData file
    adata = sc.read_h5ad(ref_path)
    
    # Ensure the cell type key exists in .obs
    if ct_key not in adata.obs.columns:
        raise ValueError(f"Key '{ct_key}' not found in .obs")
    
    # Filter the AnnData object
    filtered_adata = adata[adata.obs[ct_key].isin(cell_types_of_interest)].copy()
    
    # Create output directory if it doesn't exist
    os.makedirs(out_dir, exist_ok=True)
    
    # Define output file path
    base_name = os.path.basename(ref_path).replace(".h5ad", "_filtered.h5ad")
    out_path = os.path.join(out_dir, base_name)
    
    # Save the filtered AnnData object
    filtered_adata.write(out_path)
    
    return out_path

```
