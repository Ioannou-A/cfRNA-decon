```python
import os
import glob
import pandas as pd
import shutil

# directory with rerun files
input_dir = "Decon-COO-Results_Chalasani_NASH_nuSVR/"
pattern = os.path.join(input_dir, "nuSVR_Rerun_*.txt")

files = glob.glob(pattern)

samples = {}

for f in files:
    name = os.path.basename(f)
    parts = name.replace(".txt", "").split("_")

    # filename format: nuSVR_Rerun_<sample>_C<val>_nu<val>.txt
    sample = parts[2]
    C_val = parts[3].replace("C", "")
    nu_val = parts[4].replace("nu", "")

    df = pd.read_csv(f, sep="\t")
    rmse = float(df["RMSE-PredictedCounts"].iloc[0])

    # keep best RMSE for each sample
    if sample not in samples or rmse < samples[sample]["rmse"]:
        samples[sample] = {
            "rmse": rmse,
            "file": f,
            "C": C_val,
            "nu": nu_val
        }

# copy and rename the winning file
for sample, info in samples.items():
    best_file = info["file"]
    rmse = info["rmse"]
    C_val = info["C"]
    nu_val = info["nu"]

    new_name = f"nuSVR_CountsRMSE_Random_{sample}.txt"
    new_path = os.path.join(input_dir, new_name)

    # COPY (keeps original)
    shutil.copy(best_file, new_path)

    print(f"Sample: {sample}")
    print(f"  → Best RMSE: {rmse:.6f}")
    print(f"  → Best C:  {C_val}")
    print(f"  → Best nu: {nu_val}")
    print(f"  → Copied winning file to: {new_path}\n")
```
