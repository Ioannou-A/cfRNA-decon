This script removes the expression (zeroes it) of the faster degrading genes that are present in the simulated sample, ranging frop top 0 to 40%. After removal, the data is still normalised.

Input:
1) Counts file of the simulutated samples:
    - 1000 samples with random contributions (20250616_All-Tissues-NoDup_Random_Simulated_v2_Counts.txt) generated by the 20250618_SimulatedDatasets_v2_NoDup_1000/00_Random-Sampling_Simulated_v2_1000.py script
    - 250 samples with unirfom contributions (20250616_All-Tissues-NoDup_Uniform_Simulated_v2_Counts.txt) generated by the 20250618_SimulatedDatasets_v2_NoDup_1000/00_Uniform-Sampling_Simulated_v2_250.py script

2) List of genes with the faster degradation rate (Degradation_top_*_percent_recovered_genes.txt) generated by the Raw-Published/06_Degradation-vs-Matrix_Degradation-top-genes.ipynb script

```python
import pandas as pd
import os

# Paths
counts_file = "Data/20250616_All-Tissues-NoDup_Random_Simulated_v2_Counts.txt"
gene_lists = [
    "Data/Degradation_top_10_percent_recovered_genes.txt",
    "Data/Degradation_top_20_percent_recovered_genes.txt",
    "Data/Degradation_top_30_percent_recovered_genes.txt",
    "Data/Degradation_top_40_percent_recovered_genes.txt"
]

# Load the counts matrix
counts_df = pd.read_csv(counts_file, sep="\t", index_col=0)

# Process each gene list
for gene_list_path in gene_lists:
    # Read the gene list
    with open(gene_list_path, "r") as f:
        gene_list = [line.strip() for line in f if line.strip()]
    
    # Determine matching genes
    matching_genes = set(gene_list) & set(counts_df.index)
    print(f"{os.path.basename(gene_list_path)}: {len(matching_genes)} genes out of {len(gene_list)} matched")
    
    # Zero out expression of matching genes
    modified_df = counts_df.copy()
    modified_df.loc[list(matching_genes)] = 0

    # CPM-normalize after gene removal
    cpm_df = modified_df.div(modified_df.sum(axis=0), axis=1) * 1e6
    
    # Generate output file name
    suffix = os.path.basename(gene_list_path).replace("Degradation_", "").replace("_recovered_genes.txt", "")
    output_path = f"Data/20250616_All-Tissues-NoDup_Random_Degradation_{suffix}_removed.txt"
    
    # Save CPM-normalized matrix
    cpm_df.to_csv(output_path, sep="\t")

print("Processing complete.")
```
