This script removes the expression (zeroes it) of the faster degrading genes that are present in the pseudo-bulk sample, ranging frop top 0 to 40%. After removal, the data is still CPM normalised.

Input:
1) Counts file of the simulutated pseudo bulk samples:
    - 850 samples with random contributions (Random_v2C_All-Counts.txt) generated by the /COO-Simulated/02_Pseudo-bulk/03_PseudoSample_Generation_functions.ipynb script


2) List of genes with the faster degradation rate (Degradation_top_*_percent_recovered_genes_COO_Random_v2C.txt) generated by the Raw-Published/07_Degradation-vs-COO_Degradation-top-genes.ipynb script

```python
import pandas as pd
import os

# Paths
counts_file = "./Data/Random_v2C_All-Counts.txt"
gene_lists = [
    "./Data/Degradation_top_10_percent_recovered_genes_COO_Random_v2C.txt",
    "./Data/Degradation_top_20_percent_recovered_genes_COO_Random_v2C.txt",
    "./Data/Degradation_top_30_percent_recovered_genes_COO_Random_v2C.txt",
    "./Data/Degradation_top_40_percent_recovered_genes_COO_Random_v2C.txt"
]

# Load the counts matrix
counts_df = pd.read_csv(counts_file, sep="\t", index_col=0)

# Process each gene list
for gene_list_path in gene_lists:
    # Read the gene list
    with open(gene_list_path, "r") as f:
        gene_list = [line.strip() for line in f if line.strip()]
    
    # Determine matching genes
    matching_genes = set(gene_list) & set(counts_df.index)
    print(f"{os.path.basename(gene_list_path)}: {len(matching_genes)} genes out of {len(gene_list)} matched")
    
    # Zero out expression of matching genes
    modified_df = counts_df.copy()
    modified_df.loc[list(matching_genes)] = 0

    # CPM-normalize after gene removal
    cpm_df = modified_df.div(modified_df.sum(axis=0), axis=1) * 1e6
    
    # Generate output file name
    suffix = os.path.basename(gene_list_path).replace("Degradation_", "").replace("_recovered_genes_COO_Random_v2C.txt", "")
    output_path = f"Data/Random_v2C_{suffix}_removed.txt"
    
    # Save CPM-normalized matrix
    cpm_df.to_csv(output_path, sep="\t")

print("Processing complete.")
```
